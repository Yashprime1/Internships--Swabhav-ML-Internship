```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_excel('Analysis Data 1.xlsx')
```

```python
df
```

```python
df.describe(include='all')
```

```python
df.info()
```

```python
len(df['Products'].unique())
```

```python
len(df['Date'].unique())
```

```python
df_products = df.groupby(by=['Products','Date'], as_index=False).agg({'Sr. No.': pd.Series.nunique})
```

```python
df_products
```

```python
dates=df_products['Date'].unique()
dates=sorted(dates)
print(dates)
```

```python
d={}
for date in dates:
    d[date]=df_products[df_products['Date']==date]
```

```python
for date in dates:
    print('********************************************************************************************************************************')
    print(date)
    print('********************************************************************************************************************************')
    print(d[date])
    print(type(d[date]))
    print('********************************************************************************************************************************')
    
```

```python
# changing column names
```

```python
for date in dates:
    d[date].columns=['Products','Date','Count']
    
```

## Checking Normal Distribution through mean and median

```python
for date in dates:
    print('********************************************************************************************************************************')
    print(date)
    print('********************************************************************************************************************************')

    print('********************************************************************************************************************************')
```

## Analyzing each date seperately

```python
## date 1   ##--- assumption -- all product names will be shown by indices  (since there are many products to be shown or overlap of names would occur in x-axis)-- ##
```

```python
d[dates[0]]=d[dates[0]].sort_values(by='Count',ascending=False)
b=sns.barplot(x='Products',y='Count',data=d[dates[0]])
plt.gcf().set_size_inches(50,15)
b.axes.set_title(dates[0],fontsize=50)
b.set_xlabel('Products',fontsize=30)
b.set_ylabel('Count',fontsize=30)
_, ylabels = plt.yticks()
_, xlabels = plt.xticks()
print(ylabels[0])
print(xlabels[0])
b.set_yticklabels(b.get_yticks(), size=30)
b.set_xticklabels(b.get_xticks(), size=30)
```

```python
print(d[dates[0]].iloc[0]['Products'],d[dates[0]].iloc[0]['Count'])
print(d[dates[0]].iloc[1]['Products'],d[dates[0]].iloc[1]['Count'])
print(d[dates[0]].iloc[2]['Products'],d[dates[0]].iloc[2]['Count'])
```

```python
print('Mean : ',d[dates[0]]['Count'].mean())
print('Median : ',d[dates[0]]['Count'].median())
```
