## Common Products in both exhibitions

```python
len(list(set(df1_products['Products']) & set(df2_products['Products'])))
```

```python
common_products=list(set(df1_products['Products']) & set(df2_products['Products']))
```

```python
common_products
```

```python
print(df1_products[df1_products['Products'].isin(common_products)])
```

```python
print(df2_products[df2_products['Products'].isin(common_products)])
```

```python
d_year_18=df1_products[df1_products['Products'].isin(common_products)]
```

```python
d_year_19=df2_products[df2_products['Products'].isin(common_products)]
```

```python
df_combined=d_year_18.merge(d_year_19,on='Products',how='left')
```

```python
df_combined.columns=['Products','18_sales','19_sales']
```

```python
df_combined
```

```python
df_combined['total_sales']=df_combined.sum(axis=1)
df_combined
```

```python
df_combined['total_sales'].min()
```

```python
df_combined['total_sales'].max()
```

```python
bins=[0,10,100,300,500,800]
group=['very low sales','low sales','Average sales','high sales','very high sales']
```

```python
df_combined['SalesBin']=pd.cut(df_combined['total_sales'],bins,labels=group)
```

```python
df_combined
```

```python
df_combined.drop(columns=['total_sales'],inplace=True)
```

```python
df_combined
```

```python
df_vlow_sales_product=df_combined[df_combined['SalesBin']=='very low sales']
df_low_sales_product=df_combined[df_combined['SalesBin']=='low sales']
df_avg_sales_product=df_combined[df_combined['SalesBin']=='Average sales']
df_high_sales_product=df_combined[df_combined['SalesBin']=='high sales']
df_vhigh_sales_product=df_combined[df_combined['SalesBin']=='very high sales']
```

## Analyzing each product division as per categories
