```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
plt.isinteractive()
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
daywise_bill=df.groupby(by='day',as_index=False)['total_bill'].sum()
```

```python
daywise_bill
```

```python
plt.style.use('seaborn')
```

```python
plt.pie(daywise_bill['total_bill'],labels=daywise_bill['day'],autopct="%.2f",explode=(0,0,0,0.2))
```

 bar graph and line graph
 
 time series like sales data

```python
plt.plot(daywise_bill['day'],daywise_bill['total_bill'])
plt.title('Daywise sales')
plt.xlabel('Days')
plt.ylabel('Total_bill')
```

```python
plt.figure(figsize=(10,5))
plt.bar(daywise_bill['day'],daywise_bill['total_bill'])
plt.title('Daywise sales')
plt.xlabel('Days')
plt.ylabel('Total_bill')
```

# objects api

```python
ax=plt.subplot()
ax.bar(daywise_bill['day'],daywise_bill['total_bill'])
ax.set_title('Daywise sales')
ax.set_xlabel('Days')
ax.set_ylabel('Total_bill')
```

```python
fig,ax=plt.subplots(1,2)
fig.suptitle('Daywise sales')
ax[0].bar(daywise_bill['day'],daywise_bill['total_bill'])
ax[0].set_title('Daywise sales')
ax[0].set_xlabel('Days')
ax[0].set_ylabel('Total_bill')
ax[1].pie(daywise_bill['total_bill'],labels= daywise_bill['day'],)

ax[1].set_xlabel('Days')
ax[1].set_ylabel('Total_bill')
```
