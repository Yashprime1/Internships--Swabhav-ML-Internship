```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('persons.csv')
```

```python
df
```

```python
mw=df[df['Gender']=='Male']
fw=df[df['Gender']=='Female']
```

```python
fig,ax=plt.subplots(1,2)
ax[0].set_title('Male Weight Dist')
ax[0].boxplot(x='Weight',data=mw,showmeans=True,meanline=True)
ax[0].set_xlabel('Male')
ax[0].set_ylabel('Weights in kg')
plt.gcf().set_size_inches(15,8)
ax[1].boxplot(x='Weight',data=fw,showmeans=True,meanline=True)
ax[1].set_xlabel('Female')
ax[1].set_ylabel('Weights in kg')
plt.gcf().set_size_inches(15,8)
```

```python
df=pd.read_json("https://restcountries.eu/rest/v2/all")
```

```python
df
```

```python
df_pop=df.nlargest(10, ['population']) 
df_pop
df_pop.columns
```

```python
sns.scatterplot(y='population',x='area',data=df_pop,hue='alpha3Code')
```

```python
flights=sns.load_dataset('flights')
```

```python
flights
```

```python
groupedvalues=flights.groupby('year').sum().reset_index()
groupedvalues
```

```python
bars=plt.bar(x='year',data=groupedvalues,height='passengers',)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .005, yval)
  
```

```python
grp=df_pop.groupby(["region"],)["population"].sum()
grp=pd.DataFrame(grp)
print(grp)
```

```python
grp2=df_pop.groupby(["region"],as_index=False)["area"].sum()
grp2=pd.DataFrame(grp2)
print(grp2)
```

```python
fig,ax=plt.subplots(2,1)
ax[0].bar(grp['region'],grp['population'])
ax[0].set_title('Population by region')
ax[0].set_xlabel('Region')
ax[0].set_ylabel('Population')

ax[1].bar(grp2['region'],grp2['area'])
ax[1].set_title('Area by region')
ax[1].set_xlabel('Region')
ax[1].set_ylabel('Area')
plt.gcf().set_size_inches(15,15)
```

```python
k=grp.merge(grp2)
```

```python
sns.pairplot(k)
```
