```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
```

```python
flights.info()
```

```python
tips.info()
```

```python
# categorical plots
```

```python
sns.boxplot(x='day',y='tip',data=tips)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='day',y='tip',hue='sex',data=tips)
plt.gcf().savefig('a.png')
```

```python
sns.countplot(x='day',hue='sex',data=tips)
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.catplot(x='sex',y='tip',data=tips)
```

```python
sns.catplot(x='sex',col='day',y='tip',data=tips)
```

```python
sns.barplot(x='month',y='passengers',ci=False,data=flights)
```

```python
# distribution plots
```

```python
sns.distplot(tips['tip'],kde=False)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.rugplot(tips['tip'],height=0.5)
```

```python
sns.pairplot(tips,vars=['total_bill','tip'])
```
