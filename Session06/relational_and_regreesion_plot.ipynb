```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
tips
```

## Scatter and Line Plots

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',size='smoker',data=tips)
```

```python
sns.lineplot(x='month',y='passengers',data=flights,estimator=sum)
```

```python
sns.lineplot(x='month',y='passengers',ci=False,data=flights,estimator=sum)
```

```python
sns.lineplot(x='month',y='passengers',ci=False,data=flights,estimator=sum)
plt.gcf().set_size_inches(10,5)
```

```python
sns.lineplot(x='year',y='passengers',ci=False,data=flights,estimator=sum)
plt.gcf().set_size_inches(10,5)
```

## Relplots

```python
sns.relplot(x='total_bill',hue='smoker',y='tip',data=tips,col='sex',size='time')
```

## Regplots

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,ci=False)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,ci=False)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,ci=False,hue='smoker',col='day')
```

```python
sns.residplot(x='total_bill',y='tip',data=tips)
```
