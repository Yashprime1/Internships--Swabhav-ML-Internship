# Binning 

```python
df=pd.read_csv("marks_frame.csv",header=None,names=['name','lang','score'])
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
pd.cut(df['score'],3)
```

```python
pd.cut(df['score'],[60,70,100])
```

```python
b=pd.cut(df['score'],[35,60,70,100],labels=['ok','good','excellent'])
```

```python
b.name='Grades'
```

```python
df.join(b)
```
