```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
```

```python
racoon=misc.face()
```

```python
plt.imshow(racoon)
```

```python
r1,r2=np.vsplit(racoon,2)
```

```python
plt.imshow(r1)
```

```python
plt.imshow(r2)
```

```python
rnew=np.vstack((r2,r1))
```

```python
plt.imshow(rnew)
```

```python
racoon=misc.face()
r=racoon.copy()
print(r[0,0,:])
print(r[0,1,:])

print(r[1,0,:])

```

```python
R=r[:,:,0].copy()
G=r[:,:,1].copy()
B=r[:,:,2].copy()
r[:,:,0]=G
r[:,:,1]=R
r[:,:,2]=B
```

```python
print(r[0,0,:])
print(r[0,1,:])

print(r[1,0,:])
```
